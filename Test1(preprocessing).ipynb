```python
import os
import numpy as np
from PIL import Image
import tensorflow as tf
from sklearn.model_selection import train_test_split

# Define the path to your training data directory
train_data_dir = "/content/drive/MyDrive/Goodnotes_train/train"

# Load and preprocess a single image
# def preprocess_image(image):
#     image = tf.image.resize(image, (224, 224))
#     image = tf.image.convert_image_dtype(image, tf.float32)
#     mean = tf.constant([0.485, 0.456, 0.406])
#     std = tf.constant([0.229, 0.224, 0.225])
#     image = (image - mean) / std
#     return image
def preprocess_image(image):
    image = Image.fromarray(image).convert("RGB")  # Convert to RGB format
    image = tf.image.resize(image, (224, 224))
    image = tf.image.convert_image_dtype(image, tf.float32)
    mean = tf.constant([0.485, 0.456, 0.406])
    std = tf.constant([0.229, 0.224, 0.225])
    image = (image - mean) / std
    return image


def load_images_and_labels(data_dir):
    images = []
    labels = []

    for image_name in os.listdir(data_dir):
        if image_name.startswith('.ipynb_checkpoints'):
            continue  # Skip hidden checkpoints directory
        image_path = os.path.join(data_dir, image_name)
        image = Image.open(image_path)
        image = preprocess_image(np.array(image))
        images.append(image)
        labels.append(0)  # Assuming a single class with label 0

    return images, labels


# Load and preprocess training data
train_images, train_labels = load_images_and_labels(train_data_dir)

# Split training data into training and validation sets
train_images, val_images, train_labels, val_labels = train_test_split(
    train_images, train_labels, test_size=0.2, random_state=42
)

# Create TensorFlow data loaders
batch_size = 32
train_data = tf.data.Dataset.from_tensor_slices((train_images, train_labels))
train_data = train_data.shuffle(buffer_size=len(train_images)).batch(batch_size)

val_data = tf.data.Dataset.from_tensor_slices((val_images, val_labels))
val_data = val_data.batch(batch_size)

```
